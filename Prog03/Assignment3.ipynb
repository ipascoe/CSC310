In this assignment, we are to test the decision tree and treeviz modules in different data sets, and see how the depth of the tree affects the accuracy of the predictions formulated by the decision tree.

## Part 1: Play tennis data set

Import the necessary modules:

```python
import pandas as pd
from sklearn import tree
from treeviz import tree_print
from sklearn.metrics import accuracy_score
```

Read the play tennis csv into a data frame and display the head

```python
tennis = pd.read_csv("tennis_numeric.csv")
tennis.head()
```

Drop the unnecessary and dependent variables

```python
features_df = tennis.drop(['temperature','play'],axis=1)
features_df.head()
```

Visualize the target data

```python
target_df = pd.DataFrame(tennis['play'])
target_df.head()
```

Begin to build the decision tree

```python
dtree = tree.DecisionTreeClassifier(criterion='entropy')
dtree.fit(features_df,target_df)
```

Visualize the decision tree

```python
tree_print(dtree,features_df)
```

Create a data frame of the predicted array and compare to the target data

```python
predict_array = dtree.predict(features_df)      # produces an array of labels
predicted_labels = pd.DataFrame(predict_array)  # turn it into a DF
predicted_labels.columns = ['play']          # name the column - same name as in target!
predicted_labels.head()
```

```python
target_df.head()
```

Check to see if the decision tree is 100% accurate

```python
predicted_labels.equals(target_df)
```

```python
from sklearn.metrics import accuracy_score

print("Model accuracy is {}".format(accuracy_score(target_df, predicted_labels)))
```

The decision tree produced a perfect prediction set!

## Part 2: Our custom data set: Cracker Brands

Import the csv and display the head

```python
cracker = pd.read_csv("Cracker.csv")
cracker.head()
```

Remove useless data

```python
features_df2 = cracker.drop(['id','choice'] ,axis=1)
features_df2.head()
```

Show what the target results should look like

```python
target_df2 = pd.DataFrame(cracker['choice'])
target_df2.head()
```

Initialize decision tree

```python
dtree2 = tree.DecisionTreeClassifier(criterion='entropy')
dtree2.fit(features_df2,target_df2)
```

Visualize the decision tree

```python
tree_print(dtree2,features_df2)
```

Print predicted results and target results

```python
predict_array2 = dtree2.predict(features_df2)    
predicted_labels2 = pd.DataFrame(predict_array2) 
predicted_labels2.columns = ['choice']         
predicted_labels2.head()
```

```python
target_df2.head()
```

Tests the model's accuracy by comparing the two dataframes

```python
predicted_labels2.equals(target_df2)
```

```python
from sklearn.metrics import accuracy_score

print("Model accuracy is {}".format(accuracy_score(target_df2, predicted_labels2)))
```

The accuracy is 100%, therefore the tree did a perfect job of predicting the results

### Changing the depth of the tree
Since this is a very large data set with a lot of variables, the depth of the tree is very large. Let's try restricting the depth and see how the accuracy changes. I will go through the same process as before, just changing the parameters for the decision tree to include a maximum depth.

```python
dtree3 = tree.DecisionTreeClassifier(criterion='entropy', max_depth=16)
dtree3.fit(features_df2,target_df2)
tree_print(dtree3,features_df2)
```

```python
predict_array3 = dtree3.predict(features_df2)
predicted_labels3 = pd.DataFrame(predict_array3)
predicted_labels3.columns = ['choice'] 
predicted_labels3.head()
```

```python
target_df2.head()
```

```python
predicted_labels3.equals(target_df2)
```

```python
print("Model accuracy is {}".format(accuracy_score(target_df2, predicted_labels3)))
```

As we can see here, by restricting the depth of the tree by 10 only effected the accuracy of the results by under 4%. This can help the efficiency of the system and also have an almost negligible effect on the results.
