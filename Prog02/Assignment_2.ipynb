```python
import pandas
arrests_df = pandas.read_csv("Arrests.csv")
arrests_df
```

```python
# How many columns are there?
# How many rows are there?

rows, cols = arrests_df.shape
print("How many columns there are:", rows)
print("How many rows there are:", cols)
```

### Are the columns numerical, categorical, or both?
The columns are both numerical and categorical.

### Are there columns that provide no information?
There is one column that provides no information ('Unnamed: 0'); This provides just a constant value for all observations.

```python
# For each categorical columns compute the mode

# Released column
rel_mode = arrests_df['released'].mode()
print("Mode for column 'released':\n", rel_mode, '\n')

# Colour column
col_mode = arrests_df['colour'].mode()
print("Mode for column 'colour':\n", col_mode, '\n')

# Year column
year_mode = arrests_df['year'].mode()
print("Mode for column 'year':\n", year_mode, '\n')

# Sex column
sex_mode = arrests_df['sex'].mode()
print("Mode for column 'sex':\n", sex_mode, '\n')

# Emloyed column
empl_mode = arrests_df['employed'].mode()
print("Mode for column 'employed':\n", empl_mode, '\n')

# Citizen column
cit_mode = arrests_df['citizen'].mode()
print("Mode for column 'citizen':\n", cit_mode, '\n')
```

```python
# For each numerical column compute the mean, max and min values

# Age column
age_mean = arrests_df['age'].mean()
age_max = arrests_df['age'].max()
age_min = arrests_df['age'].min()
print("Values for the 'age' column:")
print("Mean age:", age_mean)
print("Max age:", age_max)
print("Min age:", age_min, '\n')

# Checks column
print("Values for the 'checks' column:")
checks_mean = arrests_df['checks'].mean()
checks_max = arrests_df['checks'].max()
checks_min = arrests_df['checks'].min()
print("Mean number of checks:", checks_mean)
print("Max number of checks:", checks_max)
print("Min number of checks:", checks_min)
```
