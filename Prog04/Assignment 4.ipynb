## Viewing relationship between decision tree and scatter plots above

```python
feat_crack = cracker.drop(['choice'], axis = 1)
target_crack = pd.DataFrame(cracker['choice'])
Ctree = tree.DecisionTreeClassifier(criterion= 'entropy')
Ctree.fit(feat_crack, target_crack)
tree_print(Ctree, feat_crack)
```

```python
predict_array = Ctree.predict(feat_crack)    
predicted_labels = pd.DataFrame(predict_array) 
predicted_labels.columns = ['choice']       

from sklearn.metrics import accuracy_score

print("Model accuracy is {}".format(accuracy_score(target_crack, predicted_labels)))
```

## Tree with 10% less model accuracy

```python
feat_crack = cracker.drop(['choice'], axis = 1)
target_crack = pd.DataFrame(cracker['choice'])
Ctree = tree.DecisionTreeClassifier(criterion= 'entropy', max_depth=7)
Ctree.fit(feat_crack, target_crack)
tree_print(Ctree, feat_crack)
```

```python
predict_array = Ctree.predict(feat_crack)    
predicted_labels = pd.DataFrame(predict_array) 
predicted_labels.columns = ['choice']       

from sklearn.metrics import accuracy_score

print("Model accuracy is {}".format(accuracy_score(target_crack, predicted_labels)))
```

As you can see from the first few branches of the decision tree, the trends for prediction remain the same even though the depth of the tree has been sliced by 2/3. This shows heavy correlation between the price variables and the prediction of choice.
